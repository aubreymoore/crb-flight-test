```python
import pandas as pd
import plotly.express as px
```

```python
df_mass = pd.read_csv('mass.csv')
df_mass['date'] = pd.to_datetime(df_mass['date'])
df_mass
```

```python
df_beetles = pd.read_csv('beetles.csv')
df_beetles
```

```python
df = df_beetles[['ID', 'Group', 'Sex']].merge(df_mass)
df
```

```python
fig = px.line(df[df.Sex=='F'], x="date", y="milligrams", color='ID', facet_col='Group', facet_col_wrap=2, height=1000)
fig.show()
```

```python
fig = px.line(df[df.Sex=='M'], x="date", y="milligrams", color='ID', facet_col='Group', facet_col_wrap=2, height=1000)
fig.show()
```
